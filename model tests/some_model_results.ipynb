```
##BM25
import os 
from PIL import Image
import IPython.display as display
import matplotlib
%matplotlib notebook
modes=["pairing","retrieval"]
filters=[None, "and", "or"]
tokenizers=["SentenceTokenizer", "PuctDigitRemoveTokenizer", "split"]
split = [True,False]
for m in modes:
  for f in filters:
    for t in tokenizers:
      print('Argument:\n mode:{0}  ranker:BM25  filter:{1}  tokenizer:{2}\n'.format(m,f,t))
      for info in os.popen("python test_information_retrieval.py --mode {0} --ranker BM25 --filter {1} --tokenizer {2}".format(m,f,t)).readlines():
        print(info)
      if m == 'pairing':
        display.display(Image.open("pic.png").resize((400,350)))
            
```
